```python
image_temp_path = r"Pictures\temp.png"
pdf_temp_path = r"Pictures\temp.pdf"
pdf_temp2_path = r"Pictures\temp2.pdf"

def pdfpage_to_image(pdf_path, img_path, image_page):
    pdf = pdfium.PdfDocument(pdf_path)
    page = pdf[image_page]
    with page.render(scale=4).to_pil() as image:
        image.save(img_path)
        image.close()
    pdf.close()
        

def image_to_pdf(img_path, pdf_path):
    with Image.open(img_path) as image:
 
        pdf_bytes = img2pdf.convert(image.filename)
 
        with open(pdf_path, "wb") as file:
            file.write(pdf_bytes)
            file.close()
    image.close()
   

def image_grayscale(in_path, out_path):
    with Image.open(in_path) as image:
        with ImageOps.grayscale(image) as image2:
            image2.save(out_path)
            image2.close()
    image.close()
    

def pdf_to_grayscale(path):
    # Load a document
    pdf = pdfium.PdfDocument(path)

    writer = PdfWriter()

    # Loop over pages and render
    for i in range(len(pdf)):
        print(i)
        pdfpage_to_image(path, image_temp_path, i)
        print("Pdf to image")
        image_grayscale(image_temp_path, image_temp_path)
        print("image to greyscale")
        image_to_pdf(image_temp_path, pdf_temp_path)
        print("greyscale to pdf")
        input1 = open(pdf_temp_path, "rb")
        writer.append(input1)

        input1.close()
    
    pdf.close()

    output = open(path, "wb")
    writer.write(output)
    print("write out")
    output.close()
    writer.close()
```
